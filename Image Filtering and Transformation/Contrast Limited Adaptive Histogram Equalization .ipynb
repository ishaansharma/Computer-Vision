```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
# %matplotlib inline 

def plot_gray(input_image):     
    """     
    plot grayscale image with no axis     
    """  
    # plot grayscale image with gray colormap   
    plt.imshow(input_image, cmap='gray')     
    
    # turn off axis for easier view
    plt.axis('off')
    plt.show()

def plot_hist_cdf(cdf_normalized, img):
    plt.plot(cdf_normalized, color = 'b')
    plt.hist(img.flatten(),256,[0,256], color = 'r')
    plt.xlim([0,256])
    plt.legend(('cdf','histogram'), loc = 'upper left')
    plt.show()
    

    
def main():
    # read an image 
    img = cv2.imread('pics/monalisa.jpg')
    crop_gray = cv2.cvtColor(img[100:400, 100:400], cv2.COLOR_BGR2GRAY) #increase the value of 0 to zoom in 
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8)) #contrast limited adaptive histogram equalization 
    cl1 = clahe.apply(crop_gray)
    res = np.hstack((crop_gray,cl1))
    plot_gray(res)
    
if __name__ == '__main__':
    main()
```
