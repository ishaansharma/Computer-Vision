```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
%matplotlib inline 

def _plot_cv_img(input_image):     
    """     
    Converts an image from BGR to RGB and plots     
    """   
    # change color channels order for matplotlib     
    plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))          

    # For easier view, turn off axis around image     
    plt.axis('off')
    plt.show()
    

def main():
    # read an image 
    img = cv2.imread('pics/monalisa.jpg')
    
    # create transformation matrix 
    translation_matrix = np.float32([[1,1,60],[0,2,10]])
    transformed = cv2.warpAffine(img, translation_matrix, (img.shape[1],img.shape[0]))

    # Display plot
    _plot_cv_img(transformed)

if __name__ == '__main__':
    main()
```
